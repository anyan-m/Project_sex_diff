```python
def add_sex_col(df):
    transposed = df.T
    transposed['index1'] = transposed.index
    transposed['Sex'] = transposed['index1'].apply(lambda x: sex_spl_dict.get(x))
    transposed = transposed.drop('index1', axis=1)
    return transposed
```

```python
#creates df with just female samples, needs the transposed df - 'Sex'=column
def fem(df_t):
    df_fem = df_t[df_t.Sex != 'male']
    return df_fem
```

```python
#creates df with just male samples, needs the transposed df - 'Sex'=column
def male(df_t):
    df_male = df_t[df_t.Sex != 'female']
    return df_male
```

```python
def overlap_genes_function(df):
    index_list = df.index.to_list()
    cols_list = df.columns.tolist()
    good_cols = ['Description']
    remove_leftovers = [ele for ele in index_list if ele not in overlap_lst] 
    remove_columns = [ele for ele in cols_list if ele not in good_cols] 
    df = df.drop(remove_leftovers) 
    df = df.drop(remove_columns, axis=1)
    return df
```
