**Transposes df, where sample name = index and adds 'Sex' as column**

```python
def add_sex_col(df):
    transposed = df.T
    transposed['index1'] = transposed.index
    transposed['Sex'] = transposed['index1'].apply(lambda x: sex_spl_dict.get(x))
    transposed = transposed.drop('index1', axis=1)
    return transposed
```

```python
#creates df with just female samples, needs the transposed df - 'Sex'=column
def fem(df_t):
    df_fem = df_t[df_t.Sex != 'male']
    return df_fem
```

```python
#creates df with just male samples, needs the transposed df - 'Sex'=column
def male(df_t):
    df_male = df_t[df_t.Sex != 'female']
    return df_male
```

```python
def remove_genes_less_5(df):
    df_t = df.T
    df_t = df_t.drop('zero_count', axis=1)
    df_t['zeros'] = (df_t == 0).sum(axis=1)
    rows_to_remove = df_t.zeros[df_t.zeros >= 115].index.tolist() #95% is 115.89
    return rows_to_remove#want to return list of genes with <5% expression across all samples
```

```python
def remove_rows_and_transpose(df):
    df_t = df.T
    df_t = df_t.drop('zero_count', axis=1)
    df = df_t.drop(rows_to_remove).T
    return df
```
